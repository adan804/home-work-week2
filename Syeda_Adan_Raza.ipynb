```python
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/melb_data.csv")  # Replace with your actual file

# Identify and drop columns with more than 20% missing values
missing_percent = (df.isna().sum() / len(df)) * 100
df = df.loc[:, missing_percent <= 20]

# Fill missing values in numerical columns with the median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing values in categorical columns with the most frequent value (mode)
for column in df.select_dtypes(include=['object']).columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Display the cleaned dataset
print(df.head(5))
df = pd.read_csv("/content/melb_data.csv")  # Replace with your actual file

# Identify and drop columns with more than 20% missing values
missing_percent = (df.isna().sum() / len(df)) * 100
df = df.loc[:, missing_percent <= 20]

# Fill missing values in numerical columns with the median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing values in categorical columns with the most frequent value (mode)
for column in df.select_dtypes(include=['object']).columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Display the cleaned dataset
print(df.head(5))
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/melb_data.csv")  # Replace with your actual file

# 1. Find the average price of properties in each suburb
suburb_avg_price = df.groupby("Suburb")["Price"].mean()
print("Average price per suburb:\n", suburb_avg_price, "\n")

# 2. Find the total number of properties sold in each suburb
properties_per_suburb = df.groupby("Suburb")["Price"].count()
print("Total properties sold per suburb:\n", properties_per_suburb, "\n")

# 3. Identify the suburb with the highest average price
most_expensive_suburb = suburb_avg_price.idxmax()
print("Most expensive suburb:", most_expensive_suburb)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/melb_data.csv")  # Replace with your actual file

# Save the cleaned dataset into a new CSV file
df.to_csv("cleaned_melb_data.csv", index=False)
print("Cleaned dataset saved as 'cleaned_melb_data.csv'.")

# Find the average price of properties in each suburb
suburb_avg_price = df.groupby("Suburb")["Price"].mean()

# Plot the top 10 suburbs with the highest average house prices
suburb_avg_price.head(10).plot(kind="bar", figsize=(10, 5), color='skyblue')

# Customize the plot
plt.title("Average House Prices in Top 10 Suburbs")
plt.xlabel("Suburb")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
